```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from itertools import product
plt.style.use("seaborn-v0_8")
```

```python
prices_raw = pd.read_csv("prices_round_0_day_-1.csv", sep = ";", index_col = "timestamp")
```

```python
e = prices_raw[prices_raw["product"] != "TOMATOES"]
e
```

```python
e_prices = e.bid_price_1.to_frame()
e_prices["ask_price_1"] = e.ask_price_1
e_prices["mid_price"] = e.mid_price
e_prices
```

```python
start = 0
window = 1000000
range = [start, start + window]

e_prices[["bid_price_1", "ask_price_1"]].plot(figsize = (15, 8), fontsize = 13, xlim=(range[0], range[1]))
plt.legend(fontsize = 13)
plt.show()
```
